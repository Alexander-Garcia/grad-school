```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Inputs: x, y
y = np.array([[1], [0], [0], [1]]) # Class(XNOR)
```

```python
X.shape
```

```python
model = Sequential([
    Input(shape=(X.shape[-1],)),
    Dense(2, activation="sigmoid"),
    Dense(1, activation="sigmoid")
])

model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.1), metrics=['accuracy'])
model.fit(X, y, epochs=1000)
```

```python
model.summary()
```

```python
model.get_weights()
```

## Print weights and bias of model

```python
weights = model.get_weights()
layer = 1
for i in range(0, len(weights), 2):
    print(f'Weights in {layer} layer: {weights[i]}\n')
    print(f'Bias in {layer} layer: {weights[i+1]}\n')
    layer += 1
```

## predictions

```python
prediction = model.predict(X)
```
