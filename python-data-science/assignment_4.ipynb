```python
import pandas as pd

responses = [1,2,4,4,3,5,2,1,3,3,1,4,3,3,3,2,3,3,2,5]
labels = ["mediocre", "meh", "awful", "good", "excellent"]
response_series = pd.Series(data=responses)
response_freq_series = response_series.value_counts()
percentages_series = pd.Series(data={k:v / 20 for k,v in response_freq_series.items()})
df = pd.DataFrame({"frequencies": response_freq_series.values, "percentage": percentages_series.values}, index=labels)
df.plot.bar(subplots=True)

print(f'minimum is {response_series.min()}')
print(f'max is {response_series.max()}')
print(f'range is {response_series.max() - response_series.min()}')
print(f'mean is {response_series.mean()}')
print(f'median is {response_series.median()}')
print(f'mode is {response_series.mode()}')
print(f'variance is {response_series.var():.3f}')
print(f'standard dev is {response_series.std():.3f}')
```
